```python
import time
import numpy as np
import pandas as pd
```

# Pandas

```python
# creando un pandas Series 
arreglo = np.random.randn(10)
serie = pd.Series(arreglo)
print(serie)

# agregando indices a una serie
serie = pd.Series(arreglo, index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
print(serie)

print(serie['f'])
print(serie[['a', 'b', 'c']])
```

```python
# serie con llave/valor
serie = pd.Series({'dia1': 34, 'dia2': 36, 'dia3': 43, 'dia4': 45, 'dia5': 55})
print(serie)

muestra_serie = serie[['dia1','dia3','dia5']]
print(muestra_serie)

```

```python
# creando un dataframe
datos = {
    "ventas": [31,89,23,94,87],
    "fecha": ["02/08/2022", "03/08/2022", "04/08/2022", "05/08/2022", "06/08/2022"]
    }

datos_df = pd.DataFrame(datos)

print(datos_df)

# viendo una ubicacion especifica
print(datos_df.loc[0])
print(datos_df.loc[2:4])
print(datos_df.loc[[0,2,4]])


# agregando indices a los datos
datos_df = pd.DataFrame(datos, index=['dia1', 'dia2', 'dia3', 'dia4', 'dia5'])
print(datos_df)

# viendo un elemento especifico
print(datos_df.loc['dia4'])
```
