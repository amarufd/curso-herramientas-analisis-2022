```python
datos_df_sin_na = datos_df.copy()

print(datos_df_sin_na.info())

datos_df_sin_na.dropna(inplace=True)

print(datos_df_sin_na.info())
```

```python
# mostrar nan values en datos_df
print(datos_df.isna().sum())

print(datos_df.info())
```

```python
# completando nan values en datos_df
datos_df_con_numeros = datos_df.copy()
datos_df_con_numeros['Pulse'].fillna(0, inplace=True)
print(datos_df_con_numeros.info())
```

```python
# completando nan values en datos_df con el promedio de los valores
datos_df_con_sin_na = datos_df.copy()
promedio = datos_df_con_sin_na['Calories'].mean()
print(promedio)
datos_df_con_sin_na['Calories'].fillna(promedio, inplace=True)


# completando columna Maxpulse con la moda de los valores
moda = datos_df_con_sin_na['Maxpulse'].mode()[0]
print(moda)
datos_df_con_sin_na['Maxpulse'].fillna(moda, inplace=True)

# completando la columna Pulse con la media de los valores
media = datos_df_con_sin_na['Pulse'].mean()
print(media)
datos_df_con_sin_na['Pulse'].fillna(media, inplace=True)

# completando la columna Duration con la media de los valores
media = datos_df_con_sin_na['Duration'].mean()
print(media)
datos_df_con_sin_na['Duration'].fillna(media, inplace=True)

datos_df_con_sin_na.info()
```

```python
# obteniendo el valor maximo de la columna Duration
maximo = datos_df_con_sin_na['Duration'].max()
print(maximo)

datos_trecientos = datos_df_con_sin_na['Pulse'] == 117

datos_trecientos.sum()
```

```python

datos_df_con_sin_na.loc[7, 'Duration'] = 46

for index in datos_df_con_sin_na.index:
    if datos_df_con_sin_na.loc[index, 'Duration'] == 210:
        datos_df_con_sin_na.loc[index, 'Duration'] = 120


# eliminando valores con un valor especifico
for index in datos_df_con_sin_na.index:
    if datos_df_con_sin_na.loc[index, 'Duration'] == 120:
        datos_df_con_sin_na.drop(index,inplace=True)


datos_df_con_sin_na.info()
```

```python
# viendo los duplicados
print(datos_df_con_sin_na['Pulse'].duplicated)


# eliminando duplicados
datos_df_con_sin_na['Pulse'].drop_duplicates(inplace=True)
print(datos_df_con_sin_na)

# viendo informacion de datos_df_con_sin_na
print(datos_df_con_sin_na.info())
```
